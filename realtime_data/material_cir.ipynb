```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (14,4)
```

```python
empty = pd.read_csv('material_empty.csv')
iron = pd.read_csv('material_iron.csv')
concrete = pd.read_csv('material_concrete.csv')
man = pd.read_csv('material_man.csv')
```

```python
print(empty.shape)
print(iron.shape)
print(concrete.shape)
print(man.shape)
```

```python
for i in range(150):
    empty.iloc[i][:30].plot(alpha=0.1)
```

```python
for i in range(150):
    iron.iloc[i][:30].plot(alpha=0.1)
```

```python
for i in range(150):
    concrete.iloc[i][:30].plot(alpha=0.1)
```

```python
for i in range(150):
    man.iloc[i][:30].plot(alpha=0.1)
```
